# Ejercicio t-SNE

Usamos los datos de d√≠gitos que vimos en el ejemplo. Cambia la semilla para utilizar otra muestra

```python
from sklearn.manifold import TSNE
import pandas as pd
from plotnine import *
zip_train = pd.read_csv('../datos/zip-train.csv')
muestra_dig = zip_train.sample(500, random_state=288022)
```

Prueba distintos valores de perpejlidad e iteraciones y contrasta tus resultados. Empieza con 250 iteraciones, por ejemplo

```python
tsne = TSNE(n_iter=250, perplexity=20, random_state=282)
```

```python
def graficar_digitos(datos, tsne):
    tsne_digitos = tsne.fit_transform(datos.values[:,1:257])
    dat_tsne = pd.DataFrame(tsne_digitos, columns=['X1', 'X2'])
    dat_tsne = dat_tsne.assign(digito = muestra_dig.values[:,0].astype('int'))
    (ggplot(dat_tsne, aes('X1','X2', colour='factor(digito)', label='factor(digito)')) + geom_text()).draw()
graficar_digitos(muestra_dig, tsne)
```

Ahora prueba 2000 iteraciones

```python
tsne = TSNE(n_iter=5000, perplexity=20, random_state=282)
graficar_digitos(muestra_dig, tsne)
```
